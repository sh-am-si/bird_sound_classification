```python
import json
import os
import re
import pandas as pd
import numpy as np
import pymongo
```

```python
path_info = os.path.join('./data', 'info')
path_sound = os.path.join('./data', 'sound_db')
```

```python
files = []
for filename in os.listdir(path_info):
    if filename.endswith(".json") and filename.startswith('info_'): 
        files.append(filename)

files.sort(key=lambda x: int(re.search(pattern=r"\d+", string=x).group()))
files[:25]
```

```python
d = {}
for file in files:
    ind = int(re.search(pattern=r"\d+", string=file).group())
    temp_json = json.load(open(os.path.join(path_info, file)))
    d[ind] = temp_json
```

```python
keys = list(d[1].keys())
keys
```

```python
data_list = []
for k, _d in d.items():
    temp = {'key': str(k)}
    for key in keys:
        if key in _d:
            temp[key] = str(_d[key])
        else:
            temp[key] = None
    data_list.append(temp)

data_list[:3]
```

```python
df = pd.DataFrame(data=data_list)
df.head()
```

```python
with open(os.path.join('./data', 'info.json'), mode='w') as f:
    json.dump(d, f, ensure_ascii=False, indent=4)
```
